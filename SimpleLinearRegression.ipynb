```
#Predic salary
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

year_exp=[[1],[2],[2.5],[4],[5]]
salary =[[10000],[15000],[17000],[20000],[25000]]

model=LinearRegression().fit(year_exp,salary)


plt.figure()
plt.title("Salary predictor")
plt.xlabel("YEARS OF EXP")
plt.ylabel("SALARY")
plt.plot(year_exp,salary,'*')#show the actual graph
plt.plot(year_exp,model.predict(year_exp),'--')#show the straight line

print("salary for 12 years of exp:",model.predict([[12]]))
 
```
